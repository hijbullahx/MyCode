```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate 10,000 students with random GPA
np.random.seed(42)  # For reproducibility

# Define proportions
proportions = {
    'A+': 0.15,
    'A': 0.20,
    'B': 0.30,
    'C': 0.30,
    'D': 0.05
}

# Define GPA ranges for each grade
gpa_ranges = {
    'A+': (3.57, 4.0),
    'A': (3.5, 3.57),
    'B': (3.0, 3.5),
    'C': (2.0, 2.99),
    'D': (1.2, 2.0)
}

# Generate GPA based on proportions
students = []
grade_labels = []
total_students = 10000

for grade, proportion in proportions.items():
    count = int(total_students * proportion)
    gpa_min, gpa_max = gpa_ranges[grade]
    gpas = np.random.uniform(gpa_min, gpa_max, count)
    students.extend(gpas)
    grade_labels.extend([grade] * count)

# Shuffle to mix grades
student_data = list(zip(range(1, total_students + 1), students, grade_labels))
np.random.shuffle(student_data)

# Step 2: Create DataFrame
df = pd.DataFrame(student_data, columns=["StudentID", "GPA", "Grade"])

# Step 3: Export to CSV
csv_file = "students_data.csv"
df.to_csv(csv_file, index=False)

# Step 4: Visualize Grade Distribution
grade_counts = df["Grade"].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(
    grade_counts,
    labels=grade_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired.colors
)
plt.title("Grade Distribution")
plt.show()
```

```python
from tkinter import *
from tkinter import messagebox
def submit():
    name = name_entry.get()
    email = email_entry.get()
    password = password_entry.get()
    age = age_entry.get()
    price = price_entry.get()
    print(f"Name: {name}")
    print(f"Email: {email}")
    print(f"Password: {password}")
    print(f"Age: {age}")
    print(f"Price: {price}")
    messagebox.showinfo("Submitted Data", f"Name: {name}\nEmail: {email}\nPassword: {password}\nAge: {age}\nPrice: {price}")
top = Tk()
top.geometry("300x300")
top.title("Simple Form")
Label(top, text="Name:").place(x=20, y=20)
name_entry = Entry(top)
name_entry.place(x=100, y=20)

Label(top, text="Email:").place(x=20, y=60)
email_entry = Entry(top)
email_entry.place(x=100, y=60)

Label(top, text="Password:").place(x=20, y=100)
password_entry = Entry(top, show='*')
password_entry.place(x=100, y=100)

Label(top, text="Age:").place(x=20, y=140)
age_entry = Entry(top)
age_entry.place(x=100, y=140)

Label(top, text="Price:").place(x=20, y=180)
price_entry = Entry(top)
price_entry.place(x=100, y=180)
top.mainloop()
```
